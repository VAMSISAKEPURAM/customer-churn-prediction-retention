## How did you handled the null values after farming feature table

1. Transactional Aggregations (Orders & Tickets)
If a customer has a null for total_orders, total_order_amount, or total_support_tickets, it simply means they have zero orders or zero support tickets. How to handle: Fill these with 0.

2. Numeric Features (e.g., RFM, Engagement Scores, Revenue)
For missing numerical values (like monthly_recurring_revenue or nps_score), you can either fill them with 0 (assuming lack of data means no revenue/engagement) or the median/mean of that column so you don't skew the model. How to handle: Fill with 0 or the column's median.

3. Categorical Features (e.g., Segments, Payment Methods)
For text/categorical columns (like rfm_segment or billing_cycle), you can't use numbers. How to handle: Fill with the word "Unknown".

## Feature Engineering(Created features)

1. eng_avg_order_value
How we made it: total_order_amount / total_orders
Purpose: Calculates the average amount of money a customer spends every time they make a purchase.
Why it's important: A customer might have a high total_spend, but if their average order value is rapidly dropping (e.g., they used to spend $100 per order, but now only spend $10), it is a massive red flag that they are losing interest in your product, even if their lifetime total spend still looks high.
2. eng_spend_per_product
How we made it: total_spend / num_products_owned
Purpose: Measures how much revenue we extract per product the customer uses.
Why it's important: This tells the model about "Product Stickiness." If a customer spends $1,000 across 10 different products (highly integrated into your ecosystem), they are much less likely to churn than someone who spends $1,000 on just 1 single product (who could easily switch to a single competitor).
3. eng_support_intensity
How we made it: support_tickets_30d / tenure_months
Purpose: Measures the frequency of customer frustration relative to how long they have been with you.
Why it's important: Having 3 support tickets means something completely different depending on tenure. If a 5-year customer has 3 tickets, that's normal. If a brand-new 1-month customer has 3 tickets, they are experiencing massive friction and are at very high risk of immediate churn. The raw ticket count doesn't tell the model this; the intensity ratio does.
4. eng_contract_length_months
How we made it: Converted words ("month-to-month", "annual", "biennial") into numbers (1, 12, 24).
Purpose: Transforms text into a mathematical weight of commitment.
Why it's important: If you leave it as text and use One-Hot Encoding, the model sees them as 3 completely unrelated categories. By converting them to numbers, the model understands the mathematical scale of commitment (i.e., a 24-month contract is exactly twice as long as a 12-month contract). This helps the model weigh long-term loyalty correctly.
5. eng_is_high_value (Boolean: 1 or 0)
How we made it: Flagged anyone in the top 20% of total_spend.
Purpose: Explicitly tells the model "This is a VIP customer."
Why it's important: Tree-based models (like Random Forest or XGBoost) sometimes get lost in continuous numbers. By creating a hard True/False flag for the top 20%, you force the algorithm to split the data and pay special attention to the churn patterns of your most profitable accounts.
6. eng_had_recent_support_ticket (Boolean: 1 or 0)
How we made it: Flagged anyone who had $> 0$ support tickets in the last 30 days.
Purpose: Captures the "recency effect" of customer frustration.
Why it's important: Customers don't usually churn because they had a bad experience 2 years ago; they churn because they had a bad experience yesterday. This binary flag gives the model a clear, undeniable indicator that the customer is currently in a state of friction.

## What encoding strategy did you use in your project?"
* In my Customer Churn Prediction project, I didn't use a single encoding method — I built an intelligent, adaptive encoding pipeline that automatically selects the best encoding strategy for each categorical column based on its properties. I implemented this in a class called CategoricalEncoderPipeline.

The pipeline uses 5 different encoding strategies, and the decision logic works like this:

1. Ordinal Encoding — for columns that have a natural rank order, like rfm_segment (e.g., Low → Medium → High). I used scikit-learn's OrdinalEncoder with predefined category order from the config file.

2. Target Encoding — for columns that show a strong statistical association with the target variable (churn). I measured this using Cramer's V statistic — if the Cramer's V score exceeded a threshold of 0.15, I replaced each category with the mean churn rate of that category from the training data only. This prevents data leakage.

3. One-Hot Encoding — for low cardinality columns with 10 or fewer unique values. I used scikit-learn's OneHotEncoder with handle_unknown='ignore' to safely handle unseen categories at inference time.

4. Frequency Encoding — for medium cardinality columns (11–50 unique values). I replaced each category with its relative frequency in the training data. This preserves cardinality information without the dimensionality explosion of one-hot encoding.

5. Hash Encoding — for high cardinality columns (50+ unique values). I used MD5 hashing to map categories into a fixed number of bins (100 by default). This is memory-efficient and handles unseen values gracefully.

A key production concern I addressed was data leakage prevention — all encoders are fit only on training data and then applied to the test set. The entire fitted state — target means, frequency maps, sklearn encoders — is serialized to disk using joblib so the same transformations can be reproduced at inference time without re-fitting."

 ## Question: "How did you evaluate the performance of your mode
 * In my Customer Churn Prediction project, I didn't just use a single metric like accuracy — I evaluated the model using a comprehensive suite of both statistical and business metrics, because churn is an imbalanced classification problem where accuracy alone is misleading.

Here's exactly what I computed:"

1. Statistical / Classification Metrics
Metric	Why I Used It
AUROC	Primary metric — measures overall ranking ability across all thresholds. Ideal for imbalanced datasets.
Average Precision (PR-AUC)	Better than AUROC when positive class is rare — focuses on precision-recall tradeoff
Precision	Out of customers predicted to churn, how many actually churned?
Recall	Out of all customers who actually churned, how many did we catch?
F1 Score	Harmonic mean of precision and recall — balances both
Brier Score	Measures probability calibration quality — lower is better. Tells how accurate the predicted probabilities are
2. Business Metric — Top Decile Lift
"The most important business metric I added was Top Decile Lift.

The idea is: if we sort all customers by their predicted churn probability (highest to lowest) and take the top 10%, what is the churn rate in that group compared to the overall churn rate? A lift of 3x means that group has 3x more churners than random — which directly tells the business how effective the model is for targeted retention campaigns."

python
top_decile_lift = churn_rate_in_top_10% / overall_churn_rate

3. Data Drift — PSI (Population Stability Index)
"I also implemented PSI — Population Stability Index — to detect if the score distribution shifts between training and production.

PSI < 0.1 → No shift (model is stable)
PSI 0.1–0.2 → Moderate shift (monitor closely)
PSI > 0.2 → Significant shift (retrain the model)*
This is crucial for production, because a model can have great test metrics but silently degrade over time as customer behavior changes."

4. Visualizations
"I generated two diagnostic plots and saved them as artifacts:

ROC Curve — visualizes the AUROC and the true positive rate vs false positive rate tradeoff
Calibration Curve (Reliability Diagram) — shows whether the predicted probabilities actually match observed churn rates. For example, if the model says 70% probability, do 70% of those customers actually churn?"*
5. Probability Calibration
"Before evaluation, I actually calibrated the model's probabilities using isotonic regression via scikit-learn's CalibratedClassifierCV. Raw XGBoost probabilities can be overconfident. Calibration fixes this and makes the Brier Score and Calibration Curve much better."

6. Reporting & Persistence
"All metrics are automatically saved to both a JSON file and a human-readable text report in the reports/ directory after every training run, so every experiment is fully traceable and reproducible."
Follow-up Questions & Answers
Q: Why not just use accuracy?

"Churn datasets are typically imbalanced — maybe only 10–15% of customers churn. A model that predicts 'no churn' for everyone gets 85% accuracy but is completely useless. AUROC and PR-AUC are much more honest metrics for this problem."

Q: What threshold did you use?

"The threshold was configurable via the YAML config file, defaulting to 0.5. In production, the business can tune it — a lower threshold catches more churners (higher recall) at the cost of more false positives."

Q: How did you decide the model was good enough?

"I tracked AUROC as the primary gate metric, Top Decile Lift as the business metric, and Brier Score for calibration quality. A combination of these three gives a complete picture of both statistical and real-world business performance."
