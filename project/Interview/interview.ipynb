## How did you handled the null values after farming feature table

1. Transactional Aggregations (Orders & Tickets)
If a customer has a null for total_orders, total_order_amount, or total_support_tickets, it simply means they have zero orders or zero support tickets. How to handle: Fill these with 0.

2. Numeric Features (e.g., RFM, Engagement Scores, Revenue)
For missing numerical values (like monthly_recurring_revenue or nps_score), you can either fill them with 0 (assuming lack of data means no revenue/engagement) or the median/mean of that column so you don't skew the model. How to handle: Fill with 0 or the column's median.

3. Categorical Features (e.g., Segments, Payment Methods)
For text/categorical columns (like rfm_segment or billing_cycle), you can't use numbers. How to handle: Fill with the word "Unknown".

## Feature Engineering(Created features)

1. eng_avg_order_value
How we made it: total_order_amount / total_orders
Purpose: Calculates the average amount of money a customer spends every time they make a purchase.
Why it's important: A customer might have a high total_spend, but if their average order value is rapidly dropping (e.g., they used to spend $100 per order, but now only spend $10), it is a massive red flag that they are losing interest in your product, even if their lifetime total spend still looks high.
2. eng_spend_per_product
How we made it: total_spend / num_products_owned
Purpose: Measures how much revenue we extract per product the customer uses.
Why it's important: This tells the model about "Product Stickiness." If a customer spends $1,000 across 10 different products (highly integrated into your ecosystem), they are much less likely to churn than someone who spends $1,000 on just 1 single product (who could easily switch to a single competitor).
3. eng_support_intensity
How we made it: support_tickets_30d / tenure_months
Purpose: Measures the frequency of customer frustration relative to how long they have been with you.
Why it's important: Having 3 support tickets means something completely different depending on tenure. If a 5-year customer has 3 tickets, that's normal. If a brand-new 1-month customer has 3 tickets, they are experiencing massive friction and are at very high risk of immediate churn. The raw ticket count doesn't tell the model this; the intensity ratio does.
4. eng_contract_length_months
How we made it: Converted words ("month-to-month", "annual", "biennial") into numbers (1, 12, 24).
Purpose: Transforms text into a mathematical weight of commitment.
Why it's important: If you leave it as text and use One-Hot Encoding, the model sees them as 3 completely unrelated categories. By converting them to numbers, the model understands the mathematical scale of commitment (i.e., a 24-month contract is exactly twice as long as a 12-month contract). This helps the model weigh long-term loyalty correctly.
5. eng_is_high_value (Boolean: 1 or 0)
How we made it: Flagged anyone in the top 20% of total_spend.
Purpose: Explicitly tells the model "This is a VIP customer."
Why it's important: Tree-based models (like Random Forest or XGBoost) sometimes get lost in continuous numbers. By creating a hard True/False flag for the top 20%, you force the algorithm to split the data and pay special attention to the churn patterns of your most profitable accounts.
6. eng_had_recent_support_ticket (Boolean: 1 or 0)
How we made it: Flagged anyone who had $> 0$ support tickets in the last 30 days.
Purpose: Captures the "recency effect" of customer frustration.
Why it's important: Customers don't usually churn because they had a bad experience 2 years ago; they churn because they had a bad experience yesterday. This binary flag gives the model a clear, undeniable indicator that the customer is currently in a state of friction.
