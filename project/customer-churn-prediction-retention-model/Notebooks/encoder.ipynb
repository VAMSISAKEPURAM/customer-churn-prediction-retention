```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

def encode_categorical_features(df):
    """
    Takes a dataframe and applies the correct encoding (Ordinal vs One-Hot) 
    to all categorical columns based on their inherent meaning.
    """
    # Create a copy so we don't modify the original unencoded dataframe by accident
    df_encoded = df.copy()
    
    print(f"Original Shape: {df_encoded.shape}")

    # =========================================================================
    # STEP 1: ORDINAL ENCODING
    # For columns where the categories have a clear "lowest to highest" order.
    # =========================================================================
    
    # 1. Define the exact order for each column
    tier_order = ['dormant', 'low', 'medium', 'high']
    billing_order = ['Unknown', 'monthly', 'quarterly', 'annual']
    contract_order = ['month-to-month', 'annual', 'biennial']
    contract_bill_order = ['Unknown', 'month-to-month', 'annual', 'biennial']
    # RFM standard progression from worst (Lost) to best (Champions)
    rfm_order = ['Lost', 'At Risk', 'New', 'Promising', 'Loyal', 'Champions']
    
    ordinal_cols = [
        'engagement_tier', 'billing_cycle', 'contract_type', 
        'contract_type_bill', 'rfm_segment'
    ]
    
    # 2. Initialize the encoder
    ordinal_encoder = OrdinalEncoder(categories=[
        tier_order, 
        billing_order, 
        contract_order, 
        contract_bill_order,
        rfm_order
    ])
    
    # 3. Apply the encoding
    df_encoded[ordinal_cols] = ordinal_encoder.fit_transform(df_encoded[ordinal_cols])
    print(f"Applied Ordinal Encoding to {len(ordinal_cols)} columns.")
    
    # =========================================================================
    # STEP 2: ONE-HOT ENCODING
    # For columns representing distinct groups with NO mathematical ranking.
    # =========================================================================
    
    nominal_cols = [
        'payment_method', 
        'churn_reason', 
        'retention_offer_sent',
        'payment_method_bill', 
        'invoice_delivery'
    ]
    
    # Apply One-Hot Encoding using get_dummies
    # drop_first=True prevents multicollinearity by dropping the first category
    df_encoded = pd.get_dummies(df_encoded, columns=nominal_cols, drop_first=True, dtype=int)
    
    print(f"Applied One-Hot Encoding to {len(nominal_cols)} columns.")
    print(f"New Encoded Shape: {df_encoded.shape}")
    
    return df_encoded


```

```python
# =========================================================================
# HOW TO USE IT:
# =========================================================================
# Run the function on your dataframe
df = pd.read_csv(r'D:\D_Drive\vamsi vs code\EDUKRON_FILES\Customer_churn_prediction_retenction\customer-churn-prediction-retention-model\Data\final_df.csv')
df_final = encode_categorical_features(df)
df_final.head()
```
